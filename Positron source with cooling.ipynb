```python
from array import array
import numpy as np
import pylab as pl
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import math

# Constant
mu0 = math.pi
# Solenoids on the ends
B0 = 5 # Tesla
L = 0.06 # Meter
R = 0.05 # Meter
NI = B0*L/mu0/(L/2/math.sqrt((L/2)*(L/2)+R*R))
# Solenoid in the mid
B0m = 0.5 # Tesla
Lm = 1.8 # Meter
Rm = 0.05 # Meter
NIm = B0m*Lm/mu0/(Lm/2/math.sqrt((Lm/2)*(Lm/2)+Rm*Rm))

z = array('d',[])
Bz = array('d',[])

for x in np.arange(-0.5, 2.5, 0.001):
    z.append(x)
    Bl = mu0*NI/(2*L)*((x+L/2)/math.sqrt((x+L/2)*(x+L/2)+R*R)-((x-L/2)/math.sqrt((x-L/2)*(x-L/2)+R*R)))
    Br = mu0*NI/(2*L)*((x-2+L/2)/math.sqrt((x-2+L/2)*(x-2+L/2)+R*R)-((x-2-L/2)/math.sqrt((x-2-L/2)*(x-2-L/2)+R*R)))
    Bm = mu0*NIm/(2*Lm)*((x-1+Lm/2)/math.sqrt((x-1+Lm/2)*(x-1+Lm/2)+Rm*Rm)-((x-1-Lm/2)/math.sqrt((x-1-Lm/2)*(x-1-Lm/2)+Rm*Rm)))
    Bz.append(Bl+Br+Bm)

figure(figsize=(22, 3), dpi=80)
plt.plot(z, Bz, "b")
#plt.title('Magnetic field')
plt.ylabel('Bz, T', fontsize=14)
plt.xlabel('Z, Meter', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True)
plt.axis([-0.28, 2.32, 0, 6])
plt.show()
```

```python
from array import array
import numpy as np
import pylab as pl
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import math

# Constant
mu0 = math.pi
# Solenoids on the ends
B0 = 5 # Tesla
L = 0.06 # Meter
R = 0.075 # Meter
NI = B0*L/mu0/(L/2/math.sqrt((L/2)*(L/2)+R*R))
# Solenoid in the mid
B0m = 0.5 # Tesla
Lm = 19.8 # Meter
Rm = 0.075 # Meter
NIm = B0m*Lm/mu0/(Lm/2/math.sqrt((Lm/2)*(Lm/2)+Rm*Rm))

z = array('d',[])
Bz = array('d',[])

for x in np.arange(-0.5, 20.5, 0.001):
    z.append(x)
    Bl = mu0*NI/(2*L)*((x+L/2)/math.sqrt((x+L/2)*(x+L/2)+R*R)-((x-L/2)/math.sqrt((x-L/2)*(x-L/2)+R*R)))
    Br = mu0*NI/(2*L)*((x-20+L/2)/math.sqrt((x-20+L/2)*(x-20+L/2)+R*R)-((x-20-L/2)/math.sqrt((x-20-L/2)*(x-20-L/2)+R*R)))
    Bm = mu0*NIm/(2*Lm)*((x-10+Lm/2)/math.sqrt((x-10+Lm/2)*(x-10+Lm/2)+Rm*Rm)-((x-10-Lm/2)/math.sqrt((x-10-Lm/2)*(x-10-Lm/2)+Rm*Rm)))
    Bz.append(Bl+Br+Bm)

figure(figsize=(22, 6), dpi=80)
plt.plot(z, Bz, "b")
plt.title('Magnetic field')
plt.ylabel('Bz, T')
plt.xlabel('Z, Meter')
plt.grid(True)
plt.axis([-0.28, 20.32, 0, 6])
plt.show()
```
